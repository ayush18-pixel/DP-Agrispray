```python
import subprocess

def check_gpu():
    try:
        output = subprocess.check_output("nvidia-smi --query-gpu=name --format=csv,noheader", shell=True)
        gpu_name = output.decode().strip()
        print(f"ğŸ” Detected GPU: {gpu_name}")

        if "T4" in gpu_name:
            print("âœ… You are using Tesla T4 (15GB VRAM). Training should run smoothly!")
        else:
            print("âš ï¸ Warning: You are not on a Tesla T4. You got:", gpu_name)
            print("   Training will still work, but performance/memory may differ.")
    except Exception as e:
        print("âŒ Could not detect GPU. Are you sure GPU is enabled?")
        print(e)

# Run the check
check_gpu()
```

```python
# ==========================
# STEP 1: Mount Google Drive
# ==========================
from google.colab import drive
drive.mount('/content/drive')

# ==========================
# STEP 2: Install YOLOv11 (Ultralytics)
# ==========================
!pip install ultralytics

# ==========================
# STEP 3: Import Libraries
# ==========================
from ultralytics import YOLO
import os

# ==========================
# STEP 4: Set Dataset YAML Path
# ==========================
data_yaml = "/content/datasets/yellowleafs/Yellow_Leafs-2/data.yaml"
print("âœ… Dataset YAML file:", data_yaml)

# ==========================
# STEP 5: Load YOLO11m Model
# ==========================
model = YOLO("yolo11m.pt")  # medium model

# ==========================
# STEP 6: Train YOLO11m
# ==========================
model.train(
    data=data_yaml,   # path to your dataset YAML
    epochs=50,        # adjust based on dataset size
    imgsz=640,        # image size
    batch=16,         # reduce if OOM
    device=0          # first GPU (T4)
)

# ==========================
# STEP 7: Save Best Weights & Results to Drive
# ==========================
output_folder = "/content/drive/MyDrive/YOLO_results"
os.makedirs(output_folder, exist_ok=True)

# Copy best.pt and last.pt
!cp -r runs/train/exp*/weights {output_folder}/

print(f"âœ… Training finished! Weights and results saved to {output_folder}")
```

```python
# ==========================
# STEP 1: Mount Google Drive
# ==========================
from google.colab import drive
drive.mount('/content/drive')

# ==========================
# STEP 2: Install YOLOv11 (Ultralytics)
# ==========================
!pip install ultralytics

# ==========================
# STEP 3: Import Libraries
# ==========================
from ultralytics import YOLO
import os

# ==========================
# STEP 4: Load Your Trained Model
# ==========================
weights_path = "/content/drive/MyDrive/YOLO_results/best.pt"
model = YOLO(weights_path)
print(f"âœ… Loaded model from: {weights_path}")

# ==========================
# STEP 5: Set Paths for Inference
# ==========================
# Folder with new images you want to test
input_images_folder = "/content/drive/MyDrive/YOLO_inference_images"
os.makedirs(input_images_folder, exist_ok=True)  # create if not exists

# Folder to save predictions
output_results_folder = "/content/drive/MyDrive/YOLO_inference_results"
os.makedirs(output_results_folder, exist_ok=True)

# ==========================
# STEP 6: Run Inference
# ==========================
model.predict(
    source=input_images_folder,  # folder with images or single image path
    save=True,                   # save predictions
    project=output_results_folder,  # save folder
    name="predictions",          # subfolder name
    exist_ok=True,               # overwrite if exists
    conf=0.25,                   # confidence threshold
    imgsz=640                    # image size
)

print(f"âœ… Inference done! Predictions saved to: {output_results_folder}/predictions")
```

```python
# Write the Streamlit app code into app.py
app_code = """
import streamlit as st
from ultralytics import YOLO
from PIL import Image
import tempfile
import os

# Load your trained YOLO11m model
MODEL_PATH = "/content/drive/MyDrive/YOLO_results/best.pt"
model = YOLO(MODEL_PATH)

st.title("YOLO11m Leaf Disease Detection ğŸŒ¿")
st.write("Upload an image and the model will detect diseased leaves.")

uploaded_file = st.file_uploader("Choose an image", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    img = Image.open(uploaded_file)
    st.image(img, caption="Uploaded Image", use_column_width=True)

    with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as tmp_file:
        img.save(tmp_file.name)
        tmp_path = tmp_file.name

    results = model.predict(source=tmp_path, save=False, conf=0.25, imgsz=640)
    annotated_img = results[0].plot()
    st.image(annotated_img, caption="Predictions", use_column_width=True)

    os.remove(tmp_path)
"""

# Save to app.py
with open("app.py", "w") as f:
    f.write(app_code)

print("âœ… app.py created successfully!")
```
